```python
import pandas as pd
```

```python
df=pd.read_csv('homeprices.csv')
df
```

```python
dummies = pd.get_dummies(df.town) #create dummies in town column 
dummies 
```

```python
merged= pd.concat([df,dummies], axis='columns')
merged
```

```python
final= merged.drop(['town', 'west windsor'], axis ='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X= final.drop('price', axis='columns')      #to separate input features 
                                            #here all X's are input features
X
```

```python
y= final.price
y
```

```python
#train the ML model
model.fit(X,y)
```

```python
model.predict([[2800, 1, 0]])
```

```python
model.score(X,y)
```
