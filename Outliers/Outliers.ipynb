**Detect outliers based on weight, Q1, Q3, IQR**

```python
import pandas as pd

# Load CSV in pandas dataframe
df = pd.read_csv('height_weight.csv')

# Calculate Q1, Q3, and IQR for weight
Q1_weight = df['weight'].quantile(0.25)
Q3_weight = df['weight'].quantile(0.75)
IQR_weight = Q3_weight - Q1_weight

# Calculate the range within which values are considered normal
normal_range = (Q1_weight - 1.5 * IQR_weight, Q3_weight + 1.5 * IQR_weight)

# Print the normal range
print(normal_range)
```

### IQR

```python

Q1_weight = df['weight'].quantile(0.25)
Q3_weight = df['weight'].quantile(0.75)
IQR_weight = Q3_weight - Q1_weight


print(IQR_weight)
```

### Upperlimit, lowerlimit

```python

Q1_weight = df['weight'].quantile(0.25)
Q3_weight = df['weight'].quantile(0.75)
IQR_weight = Q3_weight - Q1_weight


lower_limit = Q1_weight - 1.5 * IQR_weight
upper_limit = Q3_weight + 1.5 * IQR_weight

print("Lower Limit for Weight Outliers:", lower_limit )
print("Upper Limit for Weight Outliers:", upper_limit)
```

### Find the outlier records

```python

Q1_weight = df['weight'].quantile(0.25)
Q3_weight = df['weight'].quantile(0.75)
IQR_weight = Q3_weight - Q1_weight

lower_limit = Q1_weight - 1.5 * IQR_weight
upper_limit = Q3_weight + 1.5 * IQR_weight

weight_outliers = df[(df['weight'] < lower_limit) | (df['weight'] > upper_limit)]

print(weight_outliers)
```

**Detect outliers based on height Q1, Q3**

```python

Q1_height = df['height'].quantile(0.25)
Q3_height = df['height'].quantile(0.75)
IQR_height = Q3_height - Q1_height

normal_range_height = (Q1_height - 1.5 * IQR_height, Q3_height + 1.5 * IQR_height)

print(normal_range_height)
```

### IQR

```python
# Calculate Q1, Q3, and IQR for height
Q1_height = df['height'].quantile(0.25)
Q3_height = df['height'].quantile(0.75)
IQR_height = Q3_height - Q1_height

# Print IQR for height
print(IQR_height)
```

### Upperlimit, lowerlimit

```python

Q1_height = df['height'].quantile(0.25)
Q3_height = df['height'].quantile(0.75)
IQR_height = Q3_height - Q1_height

# Calculate upper and lower limits for outliers
lower_limit_height = Q1_height - 1.5 * IQR_height
upper_limit_height = Q3_height + 1.5 * IQR_height

# Print lower and upper limits for height
print("Lower Limit for Height Outliers:", lower_limit_height)
print("Upper Limit for Height Outliers:", upper_limit_height)
```

### Find the outlier records

```python

Q1_height = df['height'].quantile(0.25)
Q3_height = df['height'].quantile(0.75)
IQR_height = Q3_height - Q1_height


lower_limit_height = Q1_height - 1.5 * IQR_height
upper_limit_height = Q3_height + 1.5 * IQR_height

height_outliers = df[(df['height'] < lower_limit_height) | (df['height'] > upper_limit_height)]


print("Height Outlier Records:")
print(height_outliers)
```
